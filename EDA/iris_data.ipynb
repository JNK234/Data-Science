```python
setosa=data.loc[data["species"]=="setosa"]
versicolor=data.loc[data["species"]=="versicolor"]
virginica=data.loc[data["species"]=="virginica"]
#i.e dividing data into three classes
```

```python
print("Means:");
print(np.mean(setosa["petal_length"]))
print(np.mean(versicolor["petal_length"]))
print(np.mean(virginica["petal_length"]))
#The mean measures the central tendency of the feature
#By the following results we get to know that petal size of setosa is smaller and petal size of other two are comparable
print("Standard Deviation:")
print(np.std(setosa["petal_length"]))
print(np.std(versicolor["petal_length"]))
print(np.std(virginica["petal_length"]))
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(data,hue="species",size=4).map(sns.distplot,"petal_length").add_legend()
plt.show()
```

```python
#The standard deviation gives the amount of variation of values along the mean value
#But mean and standard deviation can vary in large amount if there is some error in data
```

```python
#Therefore we use median of the data sequence for finding the central tendency and does'nt get corrupted by other external 
#wrong values
print("Median:");
print(np.median(setosa["petal_length"]))
print(np.median(versicolor["petal_length"]))
print(np.median(virginica["petal_length"]))
```

```python
counts,bin_edges = np.histogram(setosa['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```
